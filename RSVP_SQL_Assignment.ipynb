```python
ps.sqldf("""
SELECT
    COUNT(CASE
            WHEN title is NULL THEN id
          END) AS title_null,
    COUNT(CASE
            WHEN year is NULL THEN id
            END) AS year_null,
    COUNT(CASE 
            WHEN worlwide_gross_income IS NULL THEN id
            END) AS income_nulls,
    COUNT(CASE
            WHEN year is NULL THEN id
            END) AS year_nulls,
    COUNT(CASE
            WHEN date_published IS NULL THEN id
            END) AS date_published_nulls,
    COUNT(CASE
            when production_company IS NULL THEN id
            END) AS prod_nulls,
    COUNT(CASE
            WHEN COUNTRY IS NULL THEN id
            END) AS country_nulls,
    COUNT(CASE
            WHEN languages IS NULL THEN id
            END) AS languages_nulls,
    COUNT(CASE
            WHEN duration IS NULL THEN id
            END) AS duration_nulls
          
    FROM movie;

""")
```

*__In solution 2 above, id in each case statement has been used as a counter to count the number of null values. Whenver there is null for a column, the id increments by 1.__*

__Notice that we do not need to check for null values in the 'id' columns as it is a pimary key__

```python
ps.sqldf("""

WITH actor_ratings AS
(
SELECT 
n.name as actor_name,
    SUM(r.total_votes) AS total_votes,
    COUNT(m.id) as movie_count,
ROUND(SUM(r.avg_rating*r.total_votes)
        /SUM(r.total_votes),2) AS actor_avg_rating
FROM
names AS n
INNER JOIN
role_mapping AS a
ON n.id=a.name_id
INNER JOIN
        movie AS m
ON a.movie_id = m.id
INNER JOIN
            ratings AS r
ON m.id=r.movie_id
WHERE category = 'actor' AND LOWER(country) like '%india%'
GROUP BY actor_name
)
SELECT *,
RANK() OVER (ORDER BY actor_avg_rating DESC, total_votes DESC) AS actor_rank
FROM
actor_ratings
WHERE movie_count>=5;

""")
```

```python
ps.sqldf("""

WITH genre_summary AS(

SELECT g.genre, ROUND(AVG(m.duration),2) AS avg_duration

FROM 
    movie AS m LEFT JOIN genre AS g
        ON m.id = g.movie_id LEFT JOIN ratings AS r
            ON m.id = r.movie_id
GROUP BY g.genre
)

SELECT *, 
ROUND(SUM(avg_duration) OVER(ORDER BY genre ROWS UNBOUNDED PRECEDING),2) AS running_total_duration,
ROUND(AVG(avg_duration)  OVER(ORDER BY genre ROWS UNBOUNDED PRECEDING),2) AS moving_avg_duration
FROM genre_summary

""")
```

```python
ps.sqldf("""

WITH top_genre AS
(
SELECT g.genre,
        COUNT(m.id),
        DENSE_RANK() OVER(ORDER BY m.id DESC) AS genre_rank
FROM movie AS m LEFT JOIN genre AS g
ON m.id = g.movie_id
GROUP BY genre
)
,

top_grossing AS
(

SELECT 
    g.genre,
    m.year,
    m.title AS movie_name,
    m.worlwide_gross_income AS worldwide_gross_income,
    RANK() OVER (PARTITION BY g.genre, m.year
                    ORDER BY CAST(REPLACE(TRIM(worlwide_gross_income),"$ ","") AS UNSIGNED)DESC) AS movie_rank

FROM movie AS m LEFT JOIN genre AS g
ON m.id = g.movie_id
WHERE g.genre IN (SELECT DISTINCT genre FROM top_genre WHERE genre_rank <=3)

)

SELECT *
FROM top_grossing
WHERE movie_rank <=5




""")
```

```python
ps.sqldf("""

WITH top_prod AS 
(
SELECT
m.production_company,
COUNT(m.id) AS movie_count,
m.languages,
DENSE_RANK() OVER (ORDER BY COUNT(m.id) DESC) AS prod_comp_rank

FROM movie AS m LEFT JOIN ratings AS r
ON m.id = r.movie_id
WHERE r.median_rating >= 8 AND m.production_company IS NOT NULL AND INSTR(languages, ',') > 0
GROUP BY m.production_company
)

SELECT *
FROM top_prod 
WHERE prod_comp_rank <=2

""")
```

```python
ps.sqldf("""

WITH actress_rank AS (
    SELECT n.name AS actress_name,
           SUM(r.total_votes) AS total_votes,
           COUNT(m.id) AS movie_count,
           ROUND(SUM(r.avg_rating*r.total_votes)/SUM(r.total_votes),2) AS actress_avg_rating
    FROM movie AS m
    LEFT JOIN genre AS g ON m.id = g.movie_id
    LEFT JOIN ratings AS r ON m.id = r.movie_id
    LEFT JOIN role_mapping AS rm ON m.id = rm.movie_id
    LEFT JOIN names AS n ON rm.name_id = n.id
    WHERE lower(g.genre) = 'drama'  and rm.category ='actress' and r.avg_rating >8
    GROUP BY n.name
)
SELECT *, DENSE_RANK() OVER(ORDER BY actress_avg_rating DESC, total_votes DESC) AS actress_rank
FROM actress_rank
LIMIT 3
""")
```

```python
ps.sqldf("""

WITH actress_ratings AS
(
SELECT 
n.name as actress_name,
    SUM(r.total_votes) AS total_votes,
    COUNT(m.id) as movie_count,
ROUND(SUM(r.avg_rating*r.total_votes)/SUM(r.total_votes),2) AS actress_avg_rating
FROM
names AS n
INNER JOIN
role_mapping AS a
ON n.id=a.name_id
INNER JOIN
        movie AS m
ON a.movie_id = m.id
INNER JOIN
            ratings AS r
ON m.id=r.movie_id
INNER JOIN
genre AS g
ON m.id=g.movie_id
WHERE category = 'actress' AND lower(g.genre) ='drama'
GROUP BY actress_name
)
SELECT *,
ROW_NUMBER() OVER (ORDER BY actress_avg_rating DESC, total_votes DESC) AS actress_rank
FROM
actress_ratings
LIMIT 3;

""")
```

```python
ps.sqldf("""

WITH top_director AS 
(
SELECT n.id AS director_id,
        n.name AS director_name,
        COUNT(m.id) AS number_of_movies,
        DENSE_RANK() OVER(ORDER BY COUNT(m.id) DESC) AS dir_rank
FROM movie AS m
LEFT JOIN ratings AS r 
ON m.id = r.movie_id
LEFT JOIN director_mapping AS d
ON m.id = d.movie_id
LEFT JOIN names AS n
ON d.name_id = n.id
WHERE director_id IS NOT NULL
GROUP BY n.id
),

movie_summary AS
(

SELECT
    n.id AS director_id,
    n.name AS director_name,
    m.id AS movie_id,
    m.date_published,
    r.avg_rating,
    r.total_votes,
    m.duration,
    LEAD(date_published) OVER (PARTITION BY n.id ORDER BY m.date_published) AS next_date_published,
   LEAD(date_published) OVER (PARTITION BY n.id ORDER BY m.date_published) AS next_date_published,
    (JULIANDAY(LEAD(date_published) OVER (PARTITION BY n.id ORDER BY m.date_published)) - JULIANDAY(date_published)) AS inter_movie_days
    FROM movie AS m
        LEFT JOIN ratings AS r 
            ON m.id = r.movie_id
                LEFT JOIN director_mapping AS d
                    ON m.id = d.movie_id
                        LEFT JOIN names AS n
                            ON d.name_id = n.id 
WHERE n.id IN (SELECT director_id FROM top_director WHERE dir_rank <=9)
)

SELECT 
    director_id,
    director_name,
    COUNT(DISTINCT movie_id) AS number_of_movies,
    ROUND(AVG(inter_movie_days),0) AS avg_inter_movie_days,
    ROUND(
        SUM(avg_rating * total_votes)/SUM(total_votes),2) AS avg_rating,
    SUM(total_votes) AS total_votes,
    MIN(avg_rating) AS min_rating,
    MAX(avg_rating) AS max_rating,
    SUM(duration) AS total_duration
FROM movie_summary
GROUP BY director_id
ORDER BY number_of_movies DESC, avg_rating DESC

""")
```
